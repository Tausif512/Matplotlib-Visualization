```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=3*x
```

```python
y
```

```python
plt.plot(x,y)
```

**Adding x,y label and line plot**

```python
plt.title('Line plot',fontsize=20)
plt.xlabel('X-Label',fontsize=15)
plt.ylabel('Y-Label',fontsize=15)
```

**CHANGING LINE AESTHETICS**

```python
plt.plot(x,y,color='r',linestyle=':', linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

**ADDING TWO LINES IN A SAME PLOT**

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color='g',linestyle=':', linewidth=2)
plt.plot(x,y2,color='y',linestyle='-', linewidth=3)
plt.title('Two Lines in a plot')
plt.xlabel('X-axis')
plt.ylabel('Y- axis')
plt.grid(True)
plt.show()
```

**lINE PLOT (ADDING SUB-PLOT)**

```python
x=np.arange(1,11)
y1=2*x
y2=2*x
plt.subplot(1,2,1) #ISKA MATLAB HAI KI 1 ROW RAHEGA AUR 2 COLUMN RAHEGA
plt.plot(x,y1,color='g',linestyle=':', linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='y',linestyle=':', linewidth=2)
plt.title('Two Lines in a plot')
plt.xlabel('X-axis')
plt.ylabel('Y- axis')
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=2*x
plt.subplot(2,2,1) #ISKA MATLAB HAI KI 1 ROW RAHEGA AUR 2 COLUMN RAHEGA
plt.plot(x,y1,color='g',linestyle=':', linewidth=2)
plt.subplot(2,2,2)
plt.plot(x,y2,color='y',linestyle=':', linewidth=2)
plt.title('Two Lines in a plot')
plt.xlabel('X-axis')
plt.ylabel('Y- axis')
plt.grid(True)
plt.show()
```

**BAR PLOT (VERTICAL)**

```python
student={'Tausif':99,'Rahul':40,'Marshal':29,'Aman':78,'Sahil':67}
```

```python
name=list(student.keys())
```

```python
name
```

```python
value=list(student.values())
```

```python
value
```

```python
plt.bar(name,value)
plt.title('Distrubution of student marks',fontsize=15)
plt.xlabel('Student Names',fontsize=15)
plt.ylabel('Student Marks',fontsize=15)
plt.grid(True)
plt.show()
```

**HORIZONTAL**

```python
plt.barh(name,value,color='r')
plt.title('Distrubution of student marks',fontsize=15)
plt.xlabel('Student Marks',fontsize=15)
plt.ylabel('Student Names',fontsize=15)
plt.grid(True)
plt.show()
```

**SCATTER PLOT**

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a)
plt.title('It is scatter plot',fontsize=15)
plt.xlabel('X-Value',fontsize=15)
plt.ylabel('Y-Value',fontsize=15)
plt.grid(True)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,a,marker='*',c='r',s=40)
plt.title('It is star scatter plot',fontsize=15)
plt.xlabel('X-Value',fontsize=15)
plt.ylabel('Y-Value',fontsize=15)
plt.grid(True)
plt.show()
```

**ADDING TWO MARKERS IN SAME PLOT**

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,3,2,6,4,8,3,8,0]
```

```python
plt.scatter(x,a,marker='*',c='r',s=100)
plt.scatter(x,b,marker='.',c='g',s=250)
plt.title('Adding two marker in the same plot',fontsize=15)
plt.xlabel('X-Value',fontsize=15)
plt.ylabel('Y-Value',fontsize=15)
plt.grid(True)
plt.show()
```

**ADDING TWO SUB-PLOT**

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='r',s=200)
plt.title('Adding two sub-plot',fontsize=10)
plt.xlabel('X-Value',fontsize=10)
plt.ylabel('Y-Value',fontsize=10)
plt.grid(True)
plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='g',s=200)
plt.title('Adding two sub-plot',fontsize=10)
plt.xlabel('X-Value',fontsize=10)
plt.ylabel('Y-Value',fontsize=10)
plt.grid(True)
plt.show()
```

**PIE CHART**

```python
fruit=['Apple','Orange','Mango','Banana']
quantity=[67,39,100,22]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

**Change Aesthetics (ADD %)**

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',
       colors=['yellow','red','pink','green'])
plt.show()
```

**DoughNut-Chart**

```python
fruit=['Apple','Orange','Mango','Banana','Guava']
quantity=[36,39,100,22,55]
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%')
plt.pie([1],colors=['pink'],radius=1)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%')
plt.pie([1],colors=['black'],radius=1)
plt.show()
```
